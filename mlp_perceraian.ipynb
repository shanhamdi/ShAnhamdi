**MEMBUAT MODEL MULTILAYER PERCEPTRON**

mengimpor kelas Sequential dan Dense dari pustaka keras.

```python
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.utils.vis_utils import plot_model
from keras.utils import plot_model
import matplotlib.pyplot as plt
import pydot
import graphviz
```

membangun model Sequential dan menambahkan tiga lapisan Dense.

```python
model = Sequential()
model.add(Dense(16, input_dim=6, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['mae'])
```

mengompilasi model dengan menentukan fungsi kerugian dan algoritma optimasi

Menampilkan model dengan tiga layer Dense.

```python
model.summary()
```

Model tersebut diatas adalah sebuah Sequential model dengan tiga layer Dense.

Layer pertama (dense_194) memiliki 16 output neurons dan menerima input dari layer sebelumnya (input layer atau hidden layer). Total parameter yang dimiliki oleh layer ini adalah 112, yang terdiri dari bobot (weight) dan bias.

Layer kedua (dense_195) juga merupakan layer Dense dengan 8 output neurons dan menerima input dari layer sebelumnya. Total parameter yang dimiliki oleh layer ini adalah 136.

Layer ketiga (dense_196) adalah layer Dense terakhir dengan 1 output neuron yang akan menghasilkan prediksi akhir. Total parameter yang dimiliki oleh layer ini adalah 9.

Total parameter (257) adalah jumlah dari parameter pada semua layer.

Menampilkan gambar arsitektur model MLP

```python
plot_model(model, show_shapes=True, show_layer_names=True)
```
